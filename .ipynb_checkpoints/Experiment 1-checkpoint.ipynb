# MERGING

```python
data1 = {
    "pokemon" : ["Bulbasaur", "Ivysaur", "Venusaur"],
    "attack" : [49,62,82]
}

data2 = {
    "pokemon" : ["Charmander", "Charmeleon", "Charizard"],
    "attack" : [52,64,84]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

mergeddf = df1.merge(df2, how='right')
mergeddf
```
